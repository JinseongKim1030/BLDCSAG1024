```python
from PIL import Image
import numpy as np

# 이미지 열기
out = Image.open('output.png').convert("RGB")
original = Image.open('image.png').convert("RGB")
mask = Image.open('mask.png').convert("RGB")

# 이미지 및 마스크를 numpy 배열로 변환
out_np = np.array(out)
original_np = np.array(original)
mask_np = np.array(mask)

# 전경 및 배경 마스크
bg_mask_np = (mask_np == 255).astype(np.float32)
fg_mask_np = (mask_np != 255).astype(np.float32)

# 알파 값 설정
alpha = 0.8

# 최종 합성 이미지 계산
out_image_np = bg_mask_np * out_np + alpha * (fg_mask_np * original_np) + (1 - alpha) * (fg_mask_np * out_np)
out_image_np = out_image_np.astype(np.uint8)  # 데이터 유형 변환

# numpy 배열을 이미지로 변환
out_image = Image.fromarray(out_image_np)

# 이미지 저장 또는 보여주기
# out_image  # 또는 out_image.save('result.png')로 저장 가능
```
